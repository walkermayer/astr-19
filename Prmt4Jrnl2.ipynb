```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000
x = np.random.poisson(100, n)

histmin = np.floor(min(x))
histmax = np.ceil(max(x))
width = (histmax - histmin) / 100

bins = np.arange(histmin, histmax, width)

plt.hist(x, bins = bins, color = 'khaki' , edgecolor = 'black')
plt.xlabel('X')
plt.ylabel('Instances of x')

plt.savefig('Prmt4Jrnl2.pdf')
plt.show()
```
