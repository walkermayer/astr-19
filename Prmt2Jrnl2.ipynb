```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)

sinx = np.sin(x)
cosx = np.cos(x)
```

```python
figure,axes = plt.subplots(1,2)

axes[0].plot(x,sinx)
axes[0].set_title('sin(x)')

axes[1].plot(x,cosx)
axes[1].set_title('cos(x)')

plt.savefig('saveprmt2jrnl2.pdf')
plt.show()
```
